```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('enrique')

def draw_circle(ax, x, y, r, color='w', alpha=1.0, linestyle='solid',
                linewidth=1.0):
    circle = plt.Circle((x, y), radius=r, color=color, alpha=alpha)
    ax.add_artist(circle)
    circle.set_facecolor('none')
    circle.set_linewidth(linewidth)
    circle.set_linestyle(linestyle)


fig = plt.figure(figsize=(8,5))
ax = plt.subplot()

data = np.load('/home/epaillasv/data/eboss_v5/galaxy_cats/\
eBOSS_LRG_clustering_NGC_v5.dat+Cap.car.npy')
data = np.asarray([i for i in data if 3000 < i[2] < 3200])
ax.scatter(data[:,0], data[:,1], s=1.0, linewidth=0)

data = np.genfromtxt('/home/epaillasv/data/eboss_v5/void_cats/\
eBOSS_LRG_clustering_NGC_v5.step0')
volfrac = np.genfromtxt('/home/epaillasv/data/eboss_v5/void_cats/\
eBOSS_LRG_clustering_NGC_v5.step0.volfrac')

ind = (3000 < data[:,2]) & (data[:,2] < 3200)
data = data[ind]
volfrac = volfrac[ind]

norm = plt.Normalize()
colors = plt.cm.jet(norm(volfrac))

for i in range(len(data)):
    draw_circle(ax=ax, x=data[i,1], y=data[i,2], r=data[i,3], color=colors[i], alpha=1.0, linewidth=3.0)

# for void in data:
#     draw_circle(ax=ax, x=void[0], y=void[1], r=void[3], color='k',
#     alpha=1.0, linewidth=3.0)


# ax.set_xlabel('x [Mpc]')
# ax.set_ylabel('y [Mpc]')
# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = np.genfromtxt('/home/epaillasv/data/eboss_v5/void_cats/\
eBOSS_LRG_clustering_NGC_v5.step0')
data = np.asarray([i for i in data if 3000 < i[2] < 3200])
npoints = 100

fig1 = plt.figure(figsize=(8,5))
ax1 = plt.subplot()

fig2 = plt.figure(figsize=(8,5))
ax2 = plt.subplot()

for void in data:
    rvoid = void[3]
    xvoid = void[0]
    yvoid = void[1]
    zvoid = void[2]

    ra = np.random.uniform(0, 2*np.pi, npoints)
    cosdec = np.random.uniform(-1, 1, npoints)
    dec = np.arccos(cosdec)
    u = np.random.uniform(0, 1, npoints)
    r = rvoid * u ** (1/3)

    x = r * np.sin(dec) * np.cos(ra) + xvoid
    y = r * np.sin(dec) * np.sin(ra) + yvoid
    z = r * np.cos(dec) + zvoid
    
    dis = np.sqrt(x**2 + y**2 + z**2)
    dec = np.arccos(z / dis)
    ra = np.arccos(x / np.sqrt(x**2 + y**2))

    ax1.scatter(x, y, s=2.0)
    ax2.scatter(ra, dec, s=2.0)

#ax1.set_xlim(-6000,6000)
#ax1.set_ylim(0, 7000)
    
plt.show()
```
